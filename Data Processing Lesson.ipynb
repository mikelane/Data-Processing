### Pandas lets us download right from a URL
Neat! Of course, if you're running experiment after experiment and need to download the data over and over again, it's probably rude to keep requesting it from the server. Download it to your machine once and use a filename instead.

The parameters of the Pandas `read_csv()` function that we care about here are `header` and `index_col`. The spambase.data file does not have a header row, so you need to make sure that `read_csv` knows this. If you don't, it'll make the first row of the data file your header. Not good.

I like to use the `index_col` parameter because it provides us with an immutable index column in our data frame that is the class of the associated row. In this case, the spambase.data csv has the class as the 57th column.

```python
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/spambase/spambase.data'
data = pd.read_csv(url, header=None, index_col=57)
```

### Pandas returns an object called a DataFrame

DataFrames have a rich API that I won't go into here. See the documentation. However, among the useful features is a method called `head()` and `tail()` which work as you'd expect. As you can see below, the data set looks like it has all the positive classifications at the top and the negative classifications at the bottom.

### Shuffling and splitting the data

Scikit-Learn provides a quick way of doing this called `train_test_split()`. You must give it the instances in the data set as the first parameter, the corresponding true values as the 2nd parameter, and what fraction ought to go into the test set. In our case the fraction is 0.5. The instances are found in the DataFrame I named `data` and to get the target values, you can access the `DataFrame.index.values` field. 

I've output the shape of the results to verify the split happened like I want.

```python
X_train, X_test, y_train, y_test = sklearn.model_selection.train_test_split(data, data.index.values, test_size=0.5)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

### Now we have training and testing sets

And you can see that they appear to be shuffled, too. In fact, they are.

```python
X_train.head()
```

```python
X_test.head()
```

### Scaling the data

Recall that you are required to scale the data based on the mean and standard deviation values of the training set. This is such a common thing to do that there is an entire class dedicated to the task in Scikit-Learn. The class is `sklearn.preprocessing.StandardScaler()`. It's got various methods, but we'll focus on the `fit()` and `transform()` methods. The `fit()` method calculates the column-wise mean and standard deviation of the set you pass to it. The `transform()` method applies those values to the data sets passed in.

```python
# Create an instance of a StandardScaler that is fit to the training data
scaler = sklearn.preprocessing.StandardScaler().fit(X_train)

# Transform the training data based on that scaler and set the index values accordingly
X_train = pd.DataFrame(scaler.transform(X_train), index=y_train)  

# Transform the testing data based on that same scaler and set the index values accordingly
X_test = pd.DataFrame(scaler.transform(X_test), index=y_test)
```

### And here is the result

Spambase data read in from a csv, split into training and testing sets, and scaled based on the training set.
